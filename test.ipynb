<h3>For the Voice to text we have:  </h3>

<h4>Local:</h4>  
Whisper / WhisperX ranges from 39 M to 1.55 B models(4.1-9.3 WER in English), uses the modesl from [https://github.com/openai/whisper](https://github.com/openai/whisper) and paper from [https://arxiv.org/abs/2212.04356](https://arxiv.org/abs/2212.04356). Based on table 8 on page 22 from the paper, Whisper Large V2 has the lowest WER at an average of 12.07, and also being the best score on 5 of the 14 tests, and at worse case is only 1.8 worse on test CallHome as compared to the Whisper Large model. As of December 2022 [https://github.com/openai/whisper/discussions/661](https://github.com/openai/whisper/discussions/661), the default large model tag will use the V2 model. 

We could use the V3, but it seems more new, and I don't want to push my luck with making it work. I could not find any published papers about it, but there seems to be some issues as seen [https://deepgram.com/learn/whisper-v3-results](https://deepgram.com/learn/whisper-v3-results), or on the github issue page where they posted [https://github.com/openai/whisper/discussions/1762#discussioncomment-7557044](https://github.com/openai/whisper/discussions/1762#discussioncomment-7557044), or here is a new issue that was created for this issue [https://github.com/openai/whisper/discussions/1836](https://github.com/openai/whisper/discussions/1836). Additionally, the model did not seem to reduce or change the model param size so their would be minimal difference in speed [https://huggingface.co/openai/whisper-large-v3#model-details](https://huggingface.co/openai/whisper-large-v3#model-details). WhisperX [https://github.com/m-bain/whisperX?tab=readme-ov-file](https://github.com/m-bain/whisperX?tab=readme-ov-file) seems to also default to V2. 


NVIDIA NeMo is another model I was looking at, but seems much more complex. From what I can tell, [https://huggingface.co/nvidia/speakerverification_en_titanet_large](https://huggingface.co/nvidia/speakerverification_en_titanet_large), [https://huggingface.co/nvidia/diar_sortformer_4spk-v1](https://huggingface.co/nvidia/diar_sortformer_4spk-v1) and other similar links, the size ranges from 7M for the Quartz 5x3 model to 120M models, and about 2.2-4.3% WER on LibriSpeech tests.  Although the paper at [https://arxiv.org/abs/2505.10879v1](https://arxiv.org/abs/2505.10879v1) seems to not support NeMo, but it looks like the TitaNet achieves 1.11-1.99% DER. One that looks interesting is Canary Qwen 2.5B with nvidia/diar_sortformer_4spk-v1

Pyannote.audio 3.1 is only for diatrization, not for audio to text. The model at [https://huggingface.co/pyannote/speaker-diarization-3.1](https://huggingface.co/pyannote/speaker-diarization-3.1) suggests the average DER is 21.54. The Github at [https://github.com/pyannote/pyannote-audio?tab=readme-ov-file](https://github.com/pyannote/pyannote-audio?tab=readme-ov-file) shows that the new community or precision models are more accurate, at 22.5(increased due to the addition of new tests, but all scores went down), and 17.18 respectfully.  This is why using WhisperX by itself may not be the best play. 

SpeechBrain [https://speechbrain.github.io/](https://speechbrain.github.io/) is the last local item. It's not really a model, more like a framework. It does offer models, and if we have time, we will test those as well. However, these are already likely worse than Whisper, it's more of a techincal tool kit. We will likely use this to clean up the audio and/or speaker splitting. However, this is for goal project work if we have free time before the end of the project. 



<h4>API:</h4>
AssemblyAI is ~$0.30/hr for the highest quality. [https://www.assemblyai.com/pricing](https://www.assemblyai.com/pricing). As DND sessions are typically 4-5 hours long, this means about $1.5 per session, which seems reasonable.  According to [https://www.assemblyai.com/research/universal-2](https://www.assemblyai.com/research/universal-2), they use a 600M model. Outside of their marketing, I found some more evidence in this paper [text](https://arxiv.org/pdf/2404.09841). Looking for quality, their paper [https://arxiv.org/abs/2501.05948v1](https://arxiv.org/abs/2501.05948v1), claims they have 2.3 WER from table 7 on page 10.

Deepgram is ~$0.948/hr for the highest quality(With basic non business packages)[https://deepgram.com/pricing](https://deepgram.com/pricing).This would be ~$4.74 per session. I don't see any published papers by them. [https://artificialanalysis.ai/speech-to-text/models/deepgram](https://artificialanalysis.ai/speech-to-text/models/deepgram) 

Rev AI is $0.20/hr [https://www.rev.ai/pricing](https://www.rev.ai/pricing), for about $1 per session. Their paper [https://arxiv.org/abs/2410.03930](https://arxiv.org/abs/2410.03930) says that the ASR model is 280M parameters, which seems to match the open source model at [https://huggingface.co/Revai/reverb-asr/tree/main](https://huggingface.co/Revai/reverb-asr/tree/main) if the model is 32 bit floating points,  with it being 2.21 GB. As for diatrization, the paper claims it is 2.2M parameters in section 3.2, whereas the model at [https://huggingface.co/Revai/reverb-diarization-v2](https://huggingface.co/Revai/reverb-diarization-v2) being 423 MB. I beieve this has to do with the model using the WavLM as described in the paper, but this makes quick verification harder. Going based on table 1 and table 2 and averaging the tests, they have a WER of ~8.52. 

Google's chirp 3 model is ~$0.96/hr [https://cloud.google.com/speech-to-text](https://cloud.google.com/speech-to-text), for about $4.80 per session. Their paper [https://arxiv.org/abs/2303.01037](https://arxiv.org/abs/2303.01037), says that the model is 2B parameters. For clarity, [https://cloud.google.com/blog/products/ai-machine-learning/bringing-power-large-models-google-clouds-speech-api](https://cloud.google.com/blog/products/ai-machine-learning/bringing-power-large-models-google-clouds-speech-api) shows that this paper regarding USM is the same one as Chirp, or a very similar model. Averaging their WER from table 3, they have a WER of 16.2. 

Amazon's Transcribe is ~$1.44 [https://aws.amazon.com/transcribe/pricing/](https://aws.amazon.com/transcribe/pricing/), for about $7.20 per session. They do not have a paper stating their exact parameters I can find. No official WER either, however, [https://artificialanalysis.ai/speech-to-text/models/aws](https://artificialanalysis.ai/speech-to-text/models/aws), gives it a WER of 14.0. 

```python
import assemblyai as aai
import os

#Set your API key (We can get one at https://www.assemblyai.com/dashboard/signup if you don't have one.)
#This is for quick and dirty stuff. You should not leave your API key here. 
aai.settings.api_key = "your-key-here"

# We should use env vars here, Like this
# In the terminal
#export ASSEMBLYAI_API_KEY="your-key-here"

#Then this will pull from the env var, not keeping it in public :
if os.environ.get("ASSEMBLYAI_API_KEY") != None:
	aai.settings.api_key = os.environ.get("ASSEMBLYAI_API_KEY")


# 2. Configure transcription with speaker diarization
config = aai.TranscriptionConfig(
	speaker_labels=True,  # Enable speaker diarization
	speakers_expected=4   # Optional: tell it roughly how many speakers (DM + players)
)

# 3. Transcribe the file
transcriber = aai.Transcriber()
#Here we are using a hard coded Audio file.
transcript = transcriber.transcribe("2014-12-03 Ep. 1.5 Here There Be Gerblins - Chapter One (Super Cut).mp3", config=config)


if transcript.status == aai.TranscriptStatus.error:
	print(f"Transcription failed: {transcript.error}")
	exit

# 4. Print results with speaker labels and timestamps
for utterance in transcript.utterances:
	print(f"Speaker {utterance.speaker}: {utterance.text}")
	print(f"  [{utterance.start/1000:.2f}s - {utterance.end/1000:.2f}s]\n")

# 5. Optional: Get full text without speaker labels
print("\n--- Full transcript ---")
print(transcript.text)
```

```python
import json

outputFile = "transcript.json"
# Save as JSON with all metadata
output = {
	"audio_duration": transcript.audio_duration,
	"status": transcript.status,
	"utterances": [
		{
			"speaker": u.speaker,
			"text": u.text,
			"start": u.start / 1000,  # Convert to seconds
			"end": u.end / 1000
		}
		for u in transcript.utterances
	],
	"full_text": transcript.text
}


with open(outputFile, "w", encoding="utf-8") as f:
	json.dump(output, f, indent=2)

print("Saved to ", outputFile)
```
