<h4>Local:</h4>  
Whisper / WhisperX ranges from 39 M to 1.55 B models(4.1-9.3 WER in English), uses the modesl from [https://github.com/openai/whisper](https://github.com/openai/whisper) and paper from [https://arxiv.org/abs/2212.04356](https://arxiv.org/abs/2212.04356). Based on table 8 on page 22 from the paper, Whisper Large V2 has the lowest WER at an average of 12.07, and also being the best score on 5 of the 14 tests, and at worse case is only 1.8 worse on test CallHome as compared to the Whisper Large model. As of December 2022 [https://github.com/openai/whisper/discussions/661](https://github.com/openai/whisper/discussions/661), the default large model tag will use the V2 model. 

We could use the V3, but it seems more new, and I don't want to push my luck with making it work. I could not find any published papers about it, but there seems to be some issues as seen [https://deepgram.com/learn/whisper-v3-results](https://deepgram.com/learn/whisper-v3-results), or on the github issue page where they posted [https://github.com/openai/whisper/discussions/1762#discussioncomment-7557044](https://github.com/openai/whisper/discussions/1762#discussioncomment-7557044), or here is a new issue that was created for this issue [https://github.com/openai/whisper/discussions/1836](https://github.com/openai/whisper/discussions/1836). Additionally, the model did not seem to reduce or change the model param size so their would be minimal difference in speed [https://huggingface.co/openai/whisper-large-v3#model-details](https://huggingface.co/openai/whisper-large-v3#model-details). WhisperX [https://github.com/m-bain/whisperX?tab=readme-ov-file](https://github.com/m-bain/whisperX?tab=readme-ov-file) seems to also default to V2. 


NVIDIA NeMo is another model I was looking at, but seems much more complex. From what I can tell, [https://huggingface.co/nvidia/speakerverification_en_titanet_large](https://huggingface.co/nvidia/speakerverification_en_titanet_large), [https://huggingface.co/nvidia/diar_sortformer_4spk-v1](https://huggingface.co/nvidia/diar_sortformer_4spk-v1) and other similar links, the size ranges from 7M for the Quartz 5x3 model to 120M models, and about 2.2-4.3% WER on LibriSpeech tests.  Although the paper at [https://arxiv.org/abs/2505.10879v1](https://arxiv.org/abs/2505.10879v1) seems to not support NeMo, but it looks like the TitaNet achieves 1.11-1.99% DER. One that looks interesting is Canary Qwen 2.5B with nvidia/diar_sortformer_4spk-v1

Pyannote.audio 3.1 is only for diatrization, not for audio to text. The model at [https://huggingface.co/pyannote/speaker-diarization-3.1](https://huggingface.co/pyannote/speaker-diarization-3.1) suggests the average DER is 21.54. The Github at [https://github.com/pyannote/pyannote-audio?tab=readme-ov-file](https://github.com/pyannote/pyannote-audio?tab=readme-ov-file) shows that the new community or precision models are more accurate, at 22.5(increased due to the addition of new tests, but all scores went down), and 17.18 respectfully.  This is why using WhisperX by itself may not be the best play. 

SpeechBrain [https://speechbrain.github.io/](https://speechbrain.github.io/) is the last local item. It's not really a model, more like a framework. It does offer models, and if we have time, we will test those as well. However, these are already likely worse than Whisper, it's more of a techincal tool kit. We will likely use this to clean up the audio and/or speaker splitting. However, this is for goal project work if we have free time before the end of the project.

Rev AI is $0.20/hr [https://www.rev.ai/pricing](https://www.rev.ai/pricing), for about $1 per session. Their paper [https://arxiv.org/abs/2410.03930](https://arxiv.org/abs/2410.03930) says that the ASR model is 280M parameters, which seems to match the open source model at [https://huggingface.co/Revai/reverb-asr/tree/main](https://huggingface.co/Revai/reverb-asr/tree/main) if the model is 32 bit floating points,  with it being 2.21 GB.
```python
# Load the files
with open("transcript.json", "r", encoding="utf-8") as f:
	transcript_data = json.load(f)

with open("speaker_mapping.json", "r", encoding="utf-8") as f:
	raw_speaker_map = json.load(f)

# We transform {"Speaker A": "Name"} -> {"A": "Name"}
# We also handle numeric cases just in case (e.g., "Speaker 1" -> "1")
clean_map = {}
for key, value in raw_speaker_map.items():
	# Remove the word "Speaker " and any trailing spaces
	clean_key = key.replace("Speaker", "").strip()
	clean_map[clean_key] = value

# Apply the mapping to the transcript
relabeled_utterances = []
for u in transcript_data["utterances"]:
	original_id = u["speaker"] # This is "A", "B", etc.
	
	# Get the name from our clean map; fallback to the ID if missing
	new_name = clean_map.get(original_id, f"Speaker {original_id}")
	
	# Create the updated object
	updated_u = u.copy()
	updated_u["speaker_name"] = new_name 
	relabeled_utterances.append(updated_u)

# Save the final JSON
transcript_data["utterances"] = relabeled_utterances
output_file = "relabeled_transcript.json"

with open(output_file, "w", encoding="utf-8") as f:
	json.dump(transcript_data, f, indent=2)

print("Mapping keys normalized and saved to ", output_file)
```

```python
#Set your API key (We can get one at https://platform.openai.com/api-keys if you don't have one.)
#This is for quick and dirty stuff. You should not leave your API key here. 
OAI_api_key = "your-key-here"

# We should use env vars here, Like this
# In the terminal
#export OPENAI_API_KEY="your-key-here"

#Then this will pull from the env var, not keeping it in public :
if os.environ.get("OPENAI_API_KEY") != None:
	OAI_api_key = os.environ.get("OPENAI_API_KEY")
	
client = OpenAI(api_key=OAI_api_key)

with open("relabeled_transcript.json", "r", encoding="utf-8") as f:
	data = json.load(f)

# Extract the full labeled conversation for the prompt
# We include timestamps so the LLM can give us accurate scene markers
full_labeled_transcript = ""
for u in data["utterances"]:
	full_labeled_transcript += f"[{u['start']:.2f}s] {u['speaker_name']}: {u['text']}\n"

# 3. The "Director" Prompt
system_instruction = """
You are a Cinematic Director. Your task is to turn a D&D transcript into a storyboard.
1. Ignore Out-of-Character (OOC) talk (rules, jokes, meta-commentary).
2. Group the narrative into 'Scenes' (8-15 seconds each).
3. For each scene, provide a highly descriptive visual prompt for a text-to-video AI.
4. Each scene must have a 'start_time' and 'end_time' based on the transcript timestamps.
5. Ignore the begining and end of the session until something happens in the world to remove player talk unrelated to the campaign.
 

Format your response ONLY as a JSON object with a list of scenes:
{
"scenes": [
	{
	"scene_number": 1,
	"start_time": 0.56,
	"end_time": 12.50,
	"location": "Dark Forest",
	"visual_prompt": "Cinematic wide shot, a group of adventurers walks through a misty dark forest, ancient oak trees, high fantasy style, 4k.",
	"narrative_summary": "The party begins their journey into the woods."
	}
]
}
"""

print(f"Segmenting {len(data['utterances'])} utterances into scenes.")

# 4. API Call
response = client.chat.completions.create(
	model="gpt-4o-mini",
	messages=[
		{"role": "system", "content": system_instruction},
		{"role": "user", "content": f"Here is the transcript:\n{full_labeled_transcript}"}
	],
	response_format={"type": "json_object"}
)

# 5. Save the Storyboard
storyboard = json.loads(response.choices[0].message.content)

stroyFile="storyboard.json" 

with open(stroyFile, "w", encoding="utf-8") as f:
	json.dump(storyboard, f, indent=2)

print(f"Storyboard created with {len(storyboard['scenes'])} scenes at file loation {stroyFile}")
```

```python
#Set your API key (We can get one at https://platform.openai.com/api-keys if you don't have one.)
#This is for quick and dirty stuff. You should not leave your API key here. 
OAI_api_key = "your-key-here"

# We should use env vars here, Like this
# In the terminal
#export OPENAI_API_KEY="your-key-here"

#Then this will pull from the env var, not keeping it in public :
if os.environ.get("OPENAI_API_KEY") != None:
	OAI_api_key = os.environ.get("OPENAI_API_KEY")
	
client = OpenAI(api_key=OAI_api_key)

with open("relabeled_transcript.json", "r", encoding="utf-8") as f:
    transcript_data = json.load(f)

with open("storyboard.json", "r", encoding="utf-8") as f:
    storyboard = json.load(f)

production_script = []

# 2. Loop through each scene identified in Stage 2
for scene in storyboard["scenes"]:
    start = scene["start_time"]
    end = scene["end_time"]
    
    # Extract only the dialogue that happens within this scene's time window
    scene_dialogue = [
        f"{u['speaker_name']}: {u['text']}" 
        for u in transcript_data["utterances"] 
        if u["start"] >= start and u["start"] <= end
    ]
    
    dialogue_block = "\n".join(scene_dialogue)
    
    # 3. Create the Screenplay Prompt
    play_prompt = f"""
    Act as a professional Screenwriter. Convert this D&D dialogue into a cinematic screenplay 'Act'.
    
    SCENE SETTING: {scene['location']}
    VISUAL STYLE: {scene['visual_prompt']}
    DIALOGUE:
    {dialogue_block}

    Provide the 'Play Act' in the following JSON format:
    {{
      "scene_number": {scene['scene_number']},
      "stage_directions": "Detailed actions of characters, lighting, and camera movement.",
      "character_actions": "Specific movements for the AI video model to capture.",
      "final_video_prompt": "A refined, high-fidelity prompt for the Video Model (Kling/Luma)."
    }}
    """

    print(f"Writing Play Act for Scene {scene['scene_number']}...")
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a master screenwriter. Return ONLY valid JSON."},
            {"role": "user", "content": play_prompt}
        ],
        response_format={"type": "json_object"}
    )
    
    # Parse and store
    scene_act = json.loads(response.choices[0].message.content)
    # Merge with original scene data for a complete record
    full_scene_data = {**scene, **scene_act}
    production_script.append(full_scene_data)

# 4. Save the Final Production Script
with open("production_script.json", "w", encoding="utf-8") as f:
    json.dump(production_script, f, indent=2)

print(f"\nProduction Script complete! {len(production_script)} scenes ready for video generation.")
```

```python
import os
import json
import fal_client # The unified API for high-end video models

# 1. Setup Fal.ai API Key (Set this in Codespaces Secrets)
os.environ["FAL_KEY"] = os.getenv("FAL_KEY")

# 2. Load the Production Script
with open("production_script.json", "r", encoding="utf-8") as f:
    production_script = json.load(f)

# 3. Select a Scene (e.g., Scene 1 for the PoC)
target_scene = production_script[0]
print(f"🎬 Generating video for Scene {target_scene['scene_number']}...")
print(f"📝 Prompt: {target_scene['final_video_prompt']}")

# 4. Define the API Callback
def on_queue_update(update):
    if isinstance(update, fal_client.InProgress):
        for log in update.logs:
            print(f"⏳ [Video Gen]: {log['message']}")

# 5. Call the Model (Luma Dream Machine Ray 2)
# This is a 'wait' call; it will hang until the video is ready (usually 2-4 mins)
result = fal_client.subscribe(
    "fal-ai/luma-dream-machine/ray-2",
    arguments={
        "prompt": target_scene["final_video_prompt"],
        "aspect_ratio": "16:9",
        "loop": False
    },
    with_logs=True,
    on_queue_update=on_queue_update,
)

# 6. Extract the Video URL
video_url = result['video']['url']
print(f"\n✅ Video Generated successfully!")
print(f"🔗 View/Download Video: {video_url}")

# 7. Update the script with the link for Stage 4 (Assembly)
target_scene["video_path"] = video_url
with open("production_script_with_videos.json", "w") as f:
    json.dump(production_script, f, indent=2)
```
